```python
import pandas as pd 
import numpy as np
```

```python
from sklearn.cluster import KMeans

Iris = pd.read_csv("Iris.csv")
print(Iris.head())
X = Iris.iloc[:,[2,3]].values
modelKM = KMeans(n_clusters=3, init='k-means++',max_iter=100,n_init=10,random_state=42)
y = modelKM.fit_predict(X)
y
```

```python
cluster = pd.DataFrame(X,columns=['Petal Length','Petal Width'])
cluster['Species'] = y
centroids = modelKM.cluster_centers_

print(cluster.head())
print(cluster.tail())
centroids
```
